```python
import pandas as pd
from tqdm import tqdm
```

```python
human_eval_data = pd.read_csv('human_evaluation_gpt4.csv') # Change this based on model to score
```

```python
human_eval_data
```

```python
# Labels for logical correctness, mathematical correctness and readability
logical_corr = {'High':1,'Medium':0.5,'Low':0}
mathematical_corr = {'Correct':1,'Incorrect':0}
readability_corr = {'High':1,'Medium':0.5,'Low':0}
```

```python
human_eval_data['Logical Correctness (Low/Medium/High)'] = human_eval_data['Logical Correctness (Low/Medium/High)'].replace(logical_corr)
human_eval_data['Mathematical correctness (Correct/Incorrect)']=human_eval_data['Mathematical correctness (Correct/Incorrect)'].replace(mathematical_corr)
human_eval_data['Readability (Low/Medium/High)']=human_eval_data['Readability (Low/Medium/High)'].replace(readability_corr)
```

```python
human_eval_data
```

```python
# Dropping the rows corresponding to ambiguous/incorrect problems
human_eval_data.drop([35,37,84,93],inplace=True)
```

```python
human_eval_data['Logical Correctness (Low/Medium/High)'] = human_eval_data['Logical Correctness (Low/Medium/High)'].apply(pd.to_numeric, errors='coerce')
human_eval_data['Mathematical correctness (Correct/Incorrect)'] = human_eval_data['Mathematical correctness (Correct/Incorrect)'].apply(pd.to_numeric, errors='coerce')
human_eval_data['Readability (Low/Medium/High)'] = human_eval_data['Readability (Low/Medium/High)'].apply(pd.to_numeric, errors='coerce')
```

```python
weighted_sum = lambda x: 0.5*x['Logical Correctness (Low/Medium/High)'] + \
                         0.3*x['Mathematical correctness (Correct/Incorrect)'] + \
                         0.2*x['Readability (Low/Medium/High)']

# Apply the function to create a new 'score' column
human_eval_data['score'] = human_eval_data.apply(weighted_sum, axis=1)
```
