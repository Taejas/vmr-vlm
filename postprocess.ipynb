```python
import pandas as pd
import numpy as np
import nltk
```

```python
# Read the tsv file containing the annotations to the 100 visual math problems
df = pd.read_csv("annotations_100.tsv", sep = '\t')
pd.set_option('display.max_colwidth', None)
display(df)
```

```python
# Annotations for a specific problem prior to postprocessing
idx = 0
print(df.iloc[idx])
```

```python
# Remove small space
df.replace(r'\\,', '', regex=True, inplace=True)

# Replace symbol for angle
df.replace('∠', 'angle ', regex=True, inplace=True)
df.replace(r'm\\angle', 'angle', regex=True, inplace=True)

# Replace symbol for degrees
df.replace('°', ' degrees', regex=True, inplace=True)
df.replace(r'\^\\circ', ' degrees', regex=True, inplace=True)

# Replace times
df.replace(r'\\times', '*', regex=True, inplace=True)

# Replace div
df.replace(r'\\div', '/', regex=True, inplace=True)

# Replace pi
df.replace('π', ' pi ', regex=True, inplace=True)

# Format fractions
df.replace(r'\\frac{([^}]*)}{([^}]*)}', r'\1/\2', regex=True, inplace=True)

# Format text
df.replace(r'\\text\{([^}]*)\}', r'\1', regex=True, inplace=True)

# Format sqrt
df.replace(r'\\sqrt\{([^}]*)\}', r'sqrt(\1)', regex=True, inplace=True)

# Format overline
df.replace(r'\\overline\{([^}]*)\}', r'\1', regex=True, inplace=True)

# Remove symbols for marking mathematical expressions
df.replace(r'\\\(', '', regex=True, inplace=True)
df.replace(r'\\\)', '', regex=True, inplace=True)
df.replace(r'\\\[', '', regex=True, inplace=True)
df.replace(r'\\\]', '', regex=True, inplace=True)
df.replace(r'\\\{', '', regex=True, inplace=True)
df.replace(r'\\\}', '', regex=True, inplace=True)

# Remove $ symbols unless they are followed by a number
df.replace(r'\$([^\s]+)\$', r'\1', regex=True, inplace=True)
df.replace(r'\$(?![0-9])', '', regex=True, inplace=True)

# Remove backslashes
df.replace(r'\\', '', regex=True, inplace=True)

# Replace tabs with single space
df.replace('\t', ' ', regex=True, inplace=True)

# Replace multiple whitespaces with single whitespace
df.replace(r'\s+', ' ', regex=True, inplace=True)

# Annotations for the problem after post-processing
print(df.iloc[idx])
```

```python
# Write postprocessed dataframe to tsv
df.to_csv('postprocessed_annotations_100.csv', index=False)
```
